```python
import pandas as pd
import numpy as np
#converting csv to dataframe
df=pd.read_csv("/workspaces/python_projects/Project_customer_purchase_behaviour/customer_shopping_behavior.csv")
#filling nan values with median based on each category
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df
```

```python
#converting column names to snake case
df.columns=df.columns.str.lower().str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df
```

```python
df['age_group']=pd.qcut(df['age'],q=4,labels=['Young Adult','Adult','Middle Aged','Senior'])
df[['age','age_group']]
```

```python
#converting purchase frequency string to integers
purchase_gap={'Fortnightly': 14,
              'Weekly' : 7,
              'Annually' : 365,
              'Bi-Weekly' : 14,
              'Quarterly' : 120,
              'Every 3 Months' : 90,
              'Monthly' : 30
              }
df['purchase_frequency_days']=df['frequency_of_purchases'].map(purchase_gap)
df[['frequency_of_purchases','purchase_frequency_days']]
```
